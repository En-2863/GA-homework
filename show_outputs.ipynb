```python
import music21
from music21 import note
from music21 import stream
from music21 import *
import os
import numpy as np
from utils.conversions import *
from utils.manipulation import *
from utils.player import play_pitches
from data.txt_to_vector import *
from data.vector_to_str import *
exp_name = 'exp_3'
num = 0

def txt2music(ms):
    score=stream.Stream([])
    id=0
    while id < 32:
        key=ms[id]

        if key=="0":
            n=note.Rest()
            n.duration.quarterLength=0.5
            id+=1

            if id==32:
                score.append(n)
                break

            while ms[id]=="-":
                n.duration.quarterLength+=0.5
                id+=1
                if id==32:
                    score.append(n)
                    break
                
            score.append(n)

        else:
            n=note.Note(key) # 创建这个音
            n.duration.quarterLength=0.5
            id+=1

            if id==32:
                score.append(n)
                break

            while ms[id]=="-":
                n.duration.quarterLength+=0.5
                id+=1
                if id==32:
                    score.append(n)
                    break
            
            score.append(n)

    return score
```

```python

# load results
population = np.load(os.path.join('experiments', exp_name, 'final_population.npy'))
```

```python
 # convert format
for k in range(40,51):    
    ms=single_vec_to_str(population[k]).split(" ")
    score=txt2music(ms)

    score.show("midi")
    score.show()
```
